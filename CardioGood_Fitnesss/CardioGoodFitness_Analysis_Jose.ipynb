```python
df_age = df.groupby(by='Product')['Age'].mean()
print('Product by mean age', df_age)
print(20*'_ ')
df_inc = df.groupby(by='Product')['Income'].mean()
print('Product by mean income',df_inc)
print(20*'_ ')
df_ed = df.groupby(by='Product')['Education'].mean()
print('Product by mean education',df_ed)
print(20*'_ ')
df_gender = df.groupby(by='Product')['Gender'].value_counts()
print('Product by gender',df_gender)
print(20*'_ ')
df_use_c = df.groupby(by='Product')['Usage'].value_counts()
print('Product by gym usage',df_use_c)
print(20*'_ ')
df_miles = df.groupby(by='Product')['Miles'].mean()
print('Product by distance',df_miles)
print(20*'_ ')
df_mar = df.groupby(by='Product')['MaritalStatus'].value_counts()
print('Product by marital status',df_mar)
print(20*'_ ')
df_fit = df.groupby(by='Product')['Fitness'].mean()
print('Product by fitness ratio',df_fit)
print(20*'_ ')
df_genfit = df.groupby(by='Gender')['Fitness'].value_counts()
print('Gender by fitness commitment',df_genfit)
print(20*'_ ')
```
