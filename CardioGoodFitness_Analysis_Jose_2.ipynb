```python
df_age = df.groupby(by='Product')['Age'].mean()
print('Product by mean age', df_age)
print(20*'_ ')
df_inc = df.groupby(by='Product')['Income'].mean()
print('Product by mean income',df_inc)
print(20*'_ ')
df_ed = df.groupby(by='Product')['Education'].mean()
print('Product by mean education',df_ed)
print(20*'_ ')
df_gender = df.groupby(by='Product')['Gender'].value_counts()
print('Product by gender',df_gender)
print(20*'_ ')
df_use_c = df.groupby(by='Product')['Usage'].value_counts()
print('Product by gym usage',df_use_c)
print(20*'_ ')
df_miles = df.groupby(by='Product')['Miles'].mean()
print('Product by distance',df_miles)
print(20*'_ ')
df_mar = df.groupby(by='Product')['MaritalStatus'].value_counts()
print('Product by marital status',df_mar)
print(20*'_ ')
df_fit = df.groupby(by='Product')['Fitness'].mean()
print('Product by fitness ratio',df_fit)
print(20*'_ ')
df_genfit = df.groupby(by='Gender')['Fitness'].value_counts()
print('Gender by fitness commitment',df_genfit)
print(20*'_ ')
```

```python
# How many people earning less than 60k run more than 150miles per week? Here we assume those buy the 2 cheaper options
df_fem = df[(df['Gender']=='Female') & (df['Income']<=60000) & (df['Miles']>150)].count()
print(df_fem)
print(20*'_ ')
df_men = df[(df['Gender']=='Male') & (df['Income']<=60000) & (df['Miles']>150)].count()
print(df_men)
```

```python
#print(df['state'].isin(['NY', 'TX']))
product = ['TM195', 'TM498']
df_cheap = df[df['Product'].isin(product)]
df_cheap.head()
```

```python
# how many people that bought the 2 cheaper options and their income is less than 60k run more than 150 miles
#those could be potential buyers for the most expensive option
df_fem = df_cheap[(df_cheap['Gender']=='Female') & (df_cheap['Income']<=60000) & (df_cheap['Miles']>150)].count()
print(df_fem)
print(20*'_ ')
df_men = df_cheap[(df_cheap['Gender']=='Male') & (df_cheap['Income']<=60000) & (df_cheap['Miles']>150)].count()
print(df_men)
```
