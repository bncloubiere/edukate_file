```python
movies = df[(df['type']=='Movie') & df['duration'].notnull()]
movies

#new df where type is movies and duration is not null
```

```python
remove_word = 'min'
movies['duration'] = movies['duration'].str.replace(remove_word, '')
movies['duration'] = movies['duration'].astype(int)
movies

#removed 'min' from the end of the duration values and converted to int
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.pie(df['type'].value_counts(), labels=df['type'].value_counts().index, autopct='%1.1f%%')
plt.title('Percentage of Movies vs TV Shows')
```

```python
sns.countplot(x = 'type', data = df)
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(x='country', data=movies, order=movies['country'].value_counts().head(10).index, palette='Set2')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
df['added_month'] = df['date_added'].dt.month
monthly_additions = df['added_month'].value_counts().sort_index()

plt.figure(figsize = (7, 6))
sns.barplot(x=monthly_additions.index, y = monthly_additions.values, palette = 'magma')
plt.title("Netflix's New Content every Month")
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
```

```python
sns.lineplot(x = 'release_year', y = 'duration', data = movies)
plt.title("Movie duration per release year")
```

```python
#A plot to show the movie duration per film year (just wanted to see the spread)
plt.figure(figsize = (20,5))
sns.boxplot(x = 'release_year', y = 'duration', data = movies)
plt.xticks(rotation=90)
plt.title('Boxplot of Movie Durations Over Release Years')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')


plt.show()
```
